# Easy 1. Инвариантность интервала (0,1) для логистического отображения

Рассмотрим одномерное логистическое отображение:

x_{n+1} = r x_n (1 - x_n),   r ∈ [0,1].

## Что нужно доказать

Если начальное значение x_0 лежит в интервале (0,1), то все последующие значения x_n тоже будут оставаться в этом интервале:

0 < x_0 < 1  ⇒  0 < x_n < 1  ∀ n ∈ ℕ.

## Подробное доказательство

Используем метод математической индукции.

**База индукции (n = 0):**
По условию задачи, 0 < x_0 < 1.

**Шаг индукции:**
Предположим, что для некоторого n выполняется 0 < x_n < 1.

Рассмотрим x_{n+1}:

Так как 0 < x_n < 1, получаем 1 - x_n ∈ (0,1),
следовательно, произведение x_n(1 - x_n) тоже положительно и строго меньше 1.

Функция x(1-x) достигает максимума при x = 1/2, где x(1-x) = 1/4.
Значит, для любого x_n ∈ (0,1) имеем:
0 < x_n(1-x_n) ≤ 1/4.

Умножаем на r ∈ [0,1]:
x_{n+1} = r x_n (1 - x_n) ∈ (0, r·1/4] ⊂ (0,1),
так как r ≤ 1.

Следовательно, 0 < x_{n+1} < 1.

По принципу математической индукции это верно для всех n ∈ ℕ.

Таким образом, интервал (0,1) является инвариантным для логистического отображения при r ∈ [0,1].

# Easy 2. Влияние параметра r на отображение

Логистическое отображение имеет вид:

f_r(x) = r x (1-x)

## Как r влияет на отображение

**Вертикальное растяжение графика:**
Увеличение r «растягивает» график по вертикали, потому что максимум функции равен r/4.

**Неподвижные точки и их устойчивость:**

- При r < 1 единственная неподвижная точка — x = 0, и она устойчива.
- При r > 1 появляется дополнительная неподвижная точка (x* = 1 - 1/r), устойчивость которой зависит от r.

**Динамический характер:**

- Малые r → затухающие колебания к неподвижной точке.
- Средние r → устойчивые циклы (периодические).
- Большие r → бифуркации и хаотическое поведение.

```python
import matplotlib.pyplot as plt

r_values = [0.3, 0.5, 0.8, 1.0]
x0 = 0.3
n = 15

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, r in enumerate(r_values):
    x = x0
    points = [(0, x0)]
    
    for i in range(1, n+1):
        x = r * x * (1 - x)
        points.append((i, x))
    
    for (n_val, x_val) in points:
        axes[idx].plot(n_val, x_val, 'ro')
    
    axes[idx].axhline(0, color='black', linestyle='--', alpha=0.7)
    axes[idx].axhline(1, color='black', linestyle='--', alpha=0.7)
    axes[idx].set_title(f'r = {r}')
    axes[idx].set_xlabel('n')
    axes[idx].set_ylabel('xn')
    axes[idx].grid(True)
    axes[idx].set_ylim(-0.1, 1.1)

plt.suptitle(f'Логистическое отображение: xn+1 = r·xn·(1-xn), x0 = {x0}', fontsize=14)
plt.tight_layout()
plt.show()
```

# Анализ отображения N=2

## Вариант
**N = 2**  
**Отображение:** \( x_{n+1} = r x_n (1 - x_n)^2 \)  
**Диапазон r:** от 0 до 6.75

## 1. Свойства функции

### 1.1. Форма функции
Функция \( f(x) = r x (1 - x)^2 \) имеет кубическую форму (степень 3).

### 1.2. Нули функции
Функция равна нулю при:
- \( x = 0 \)
- \( x = 1 \)

### 1.3. Максимум функции
Максимум функции достигается при \( x = 1/3 \).  
Значение максимума равно \( 4r/27 \).

### 1.4. Диапазон параметра r
Для того чтобы значения функции оставались между 0 и 1, необходимо:
- Максимум функции ≤ 1
- \( 4r/27 ≤ 1 \)
- \( r ≤ 27/4 = 6.75 \)

## 2. Неподвижные точки

### 2.1. Уравнение
Неподвижные точки удовлетворяют: \( x = r x (1 - x)^2 \)

### 2.2. Решения
1. \( x = 0 \) (всегда неподвижная точка)
2. \( x^* = 1 - 1/\sqrt{r} \) (появляется при \( r ≥ 1 \))

## 3. Сравнение с логистическим отображением

### Логистическое отображение
\( x_{n+1} = r x_n (1 - x_n) \)

### Сходства
1. Оба имеют форму "колокола"
2. Оба равны 0 при x=0 и x=1
3. Оба используют параметр r
4. Оба имеют неподвижную точку x=0

### Различия
1. **Степень**: логистическое - квадратичное, наше - кубическое
2. **Максимум**: 
   - Логистическое: при x=0.5, значение r/4
   - Наше: при x=1/3, значение 4r/27
3. **Диапазон r**:
   - Логистическое: [0, 4]
   - Наше: [0, 6.75]
4. **Ненулевая точка**:
   - Логистическое: \( 1 - 1/r \)
   - Наше: \( 1 - 1/\sqrt{r} \)

## 4. Выводы

### 4.1. Общие выводы
- Оба отображения моделируют системы с ограниченным ростом
- Наше отображение имеет более сильное ограничение роста
- Квадрат (1-x)² означает усиленное подавление при высокой плотности

### 4.2. Биологическая интерпретация
- Логистическое: линейная конкуренция за ресурсы
- Наше: квадратичная конкуренция (территория + ресурсы)

### 4.3. Математические особенности
- Кубическая форма → более сложная динамика
- Больший диапазон r → больше возможностей для исследования
- Другие условия устойчивости

# Normal 1: Анализ неподвижных точек логистического отображения

## Логистическое отображение
x_{n+1} = r x_n (1 - x_n)

## 1.1 Неподвижные точки
Неподвижные точки находятся из уравнения:
x* = r x* (1 - x*)

Решения:
1. x* = 0 (существует всегда)
2. x* = 1 - 1/r (существует при r >= 1)

## 1.2 Количество неподвижных точек в зависимости от r

При 0 <= r < 1:
- Существует только x* = 0
- Одна неподвижная точка

При r = 1:
- Обе формулы дают x* = 0
- Одна неподвижная точка (двойная)

При r > 1:
- Существуют обе точки: x* = 0 и x* = 1 - 1/r
- Две неподвижные точки

## 1.3 Максимальное количество неподвижных точек
Максимальное количество: 2

Обоснование:
- Уравнение x = r x (1 - x) квадратное
- Квадратное уравнение имеет максимум 2 корня
- Больше 2 корней невозможно

# Normal 2: Анализ монотонности и предела логистического отображения

## Условие
Для логистического отображения:
x_{n+1} = r x_n (1 - x_n)

При условиях:
- x₀ ∈ (0; 1)
- r ∈ (0; 1]

## 1. Доказательство монотонного убывания

### 1.1 Сравнение x_{n+1} и x_n
Рассмотрим разность:
x_{n+1} - x_n = r x_n (1 - x_n) - x_n = x_n [r(1 - x_n) - 1]

### 1.2 Анализ знака разности
При x_n ∈ (0; 1) и r ∈ (0; 1]:
1. x_n > 0 (по условию)
2. r(1 - x_n) - 1 ≤ r - 1 ≤ 0 (так как r ≤ 1)

Следовательно:
x_{n+1} - x_n = x_n [r(1 - x_n) - 1] ≤ 0

### 1.3 Результат
Для всех n ≥ 0:
x_{n+1} ≤ x_n

Последовательность {x_n} монотонно убывает.

## 2. Существование предела

### 2.1 Ограниченность последовательности
1. По условию x₀ ∈ (0; 1)
2. Из доказательства Easy части: при r ∈ [0; 1] и 0 < x₀ < 1, все x_n ∈ (0; 1)
3. Последовательность ограничена снизу 0 и сверху 1

### 2.2 Монотонность
Из части 1: последовательность монотонно убывает

### 2.3 Теорема о монотонной сходимости
Монотонная ограниченная последовательность имеет предел.

### 2.4 Значение предела
Так как последовательность монотонно убывает и ограничена снизу 0, то:
lim_{n→∞} x_n = L ≥ 0

Из рекуррентного соотношения:
L = r L (1 - L)

Решения:
1. L = 0
2. L = 1 - 1/r (но при r ≤ 1, эта точка не лежит в (0; 1) кроме r = 1)

Таким образом, при r ∈ (0; 1]:
lim_{n→∞} x_n = 0

## 3. Графическая иллюстрация

Последовательность монотонно убывает к 0 при r ≤ 1.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    """Логистическое отображение"""
    return r * x * (1 - x)

r_values = [0.3, 0.5, 0.7, 1.0]
x0 = 0.8
n_steps = 20

plt.figure(figsize=(12, 8))

for i, r in enumerate(r_values, 1):
    plt.subplot(2, 2, i)
    
    trajectory = [x0]
    x = x0
    for _ in range(n_steps):
        x = logistic_map(x, r)
        trajectory.append(x)
    
    plt.plot(range(n_steps + 1), trajectory, 'bo-', linewidth=2, markersize=4)
    
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.5, label='Предел: 0')
    
    plt.title(f'r = {r}, x₀ = {x0}')
    plt.xlabel('n (шаг)')
    plt.ylabel('x_n')
    plt.grid(True, alpha=0.3)
    plt.ylim(-0.05, 0.85)
    
    plt.text(0.5, 0.9, f'Монотонное убывание', 
             transform=plt.gca().transAxes, ha='center',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))

plt.tight_layout()
plt.show()
```

# Normal 3: Анализ монотонности подпоследовательностей при r ∈ (2; 3)

## Условие
Для логистического отображения:
x_{n+1} = r x_n (1 - x_n)

При условиях:
- r ∈ (2; 3)
- x_{2n} > x* (чётные члены больше неподвижной точки)
- x_{2n+1} < x* (нечётные члены меньше неподвижной точки)
- где x* = 1 - 1/r (ненулевая неподвижная точка)

## 1. Анализ поведения функции f(x) = r x (1 - x)

### 1.1 Свойства функции при r ∈ (2; 3)
1. Неподвижная точка: x* = 1 - 1/r ∈ (1/2, 2/3)
2. Производная: f'(x) = r(1 - 2x)
3. Значение производной в x*:
   f'(x*) = r(1 - 2(1 - 1/r)) = r(1 - 2 + 2/r) = r(-1 + 2/r) = -r + 2
   Так как r ∈ (2; 3), то f'(x*) ∈ (-1; 0)

### 1.2 Поведение функции относительно x*
- f(x) - x* = r x (1 - x) - (1 - 1/r)
- Для x > x*: f(x) < x* (так как производная отрицательна)
- Для x < x*: f(x) > x* (так как производная отрицательна)

## 2. Монотонность подпоследовательности {x_{2n}} (чётные)

### 2.1 Рассмотрим x_{2n} и x_{2n+2}
x_{2n+2} = f(f(x_{2n})) = f²(x_{2n})

### 2.2 Анализ второй итерации
Так как x_{2n} > x* и f'(x*) ∈ (-1; 0), то:
1. f(x_{2n}) = x_{2n+1} < x* (по свойству функции)
2. f(x_{2n+1}) = f(f(x_{2n})) > x* (так как x_{2n+1} < x*)

### 2.3 Сравнение x_{2n} и x_{2n+2}
Рассмотрим функцию g(x) = f(f(x)) - x
При r ∈ (2; 3) и x > x*:
- g'(x*) = [f'(x*)]² ∈ (0; 1)
- g(x) монотонно возрастает в окрестности x*
- Так как x_{2n} > x* и функция возвращает значение > x*, получаем x_{2n+2} < x_{2n}

### 2.4 Результат для чётных членов
Подпоследовательность {x_{2n}} монотонно убывает:
x_{2n+2} < x_{2n}

## 3. Монотонность подпоследовательности {x_{2n+1}} (нечётные)

### 3.1 Рассмотрим x_{2n+1} и x_{2n+3}
x_{2n+3} = f(f(x_{2n+1}))

### 3.2 Анализ
Так как x_{2n+1} < x* и f'(x*) ∈ (-1; 0), то:
1. f(x_{2n+1}) = x_{2n+2} > x*
2. f(x_{2n+2}) = f(f(x_{2n+1})) < x*

### 3.3 Сравнение x_{2n+1} и x_{2n+3}
При x < x* и r ∈ (2; 3):
- f(f(x)) > x
- x_{2n+3} > x_{2n+1}

### 3.4 Результат для нечётных членов
Подпоследовательность {x_{2n+1}} монотонно возрастает:
x_{2n+3} > x_{2n+1}

## 4. Итог

1. **Подпоследовательность {x_{2n}}** (чётные):
   - Начинается выше x*
   - Монотонно убывает
   - Ограничена снизу x*

2. **Подпоследовательность {x_{2n+1}}** (нечётные):
   - Начинается ниже x*
   - Монотонно возрастает
   - Ограничена сверху x*

3. **Обе подпоследовательности** сходятся к x*:
   - lim_{n→∞} x_{2n} = x* (сверху)
   - lim_{n→∞} x_{2n+1} = x* (снизу)

4. **Общая последовательность {x_n}**:
   - Осциллирует вокруг x*
   - Сходится к x* колебательным образом

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    """Логистическое отображение"""
    return r * x * (1 - x)

def analyze_subsequences(r, x0, n_steps=30):
    """Анализ чётных и нечётных подпоследовательностей"""
    # Вычисляем неподвижную точку
    x_star = 1 - 1/r
    
    trajectory = [x0]
    x = x0
    for _ in range(n_steps):
        x = logistic_map(x, r)
        trajectory.append(x)
    
    # Разделяем на чётные и нечётные
    even_indices = list(range(0, n_steps+1, 2))
    odd_indices = list(range(1, n_steps+1, 2))
    
    even_vals = [trajectory[i] for i in even_indices]
    odd_vals = [trajectory[i] for i in odd_indices]
    
    return trajectory, even_vals, odd_vals, even_indices, odd_indices, x_star

r = 2.5  # r ∈ (2, 3)
x0 = 0.8  # x0 > x*
x_star = 1 - 1/r


trajectory, even_vals, odd_vals, even_idx, odd_idx, x_star = analyze_subsequences(r, x0)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))

ax1.plot(range(len(trajectory)), trajectory, 'bo-', linewidth=1.5, markersize=5, label='Последовательность {x_n}')
ax1.axhline(y=x_star, color='r', linestyle='--', alpha=0.7, linewidth=2, label=f'x* = {x_star:.4f}')
ax1.set_xlabel('n')
ax1.set_ylabel('x_n')
ax1.set_title(f'Логистическое отображение: r = {r}, x₀ = {x0}')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_ylim(0, 1)

ax2.plot(even_idx, even_vals, 'go-', linewidth=2, markersize=6, label='Чётные {x_{2n}}')
ax2.axhline(y=x_star, color='r', linestyle='--', alpha=0.7, linewidth=2, label=f'x*')
ax2.set_xlabel('Индекс n')
ax2.set_ylabel('x_{2n}')
ax2.set_title(f'Чётная подпоследовательность (монотонно убывает)')
ax2.grid(True, alpha=0.3)
ax2.legend()

is_decreasing = all(even_vals[i] > even_vals[i+1] for i in range(len(even_vals)-1))
ax2.text(0.02, 0.98, f'Монотонно убывает: {is_decreasing}', 
         transform=ax2.transAxes, va='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen' if is_decreasing else 'pink', alpha=0.8))

ax3.plot(odd_idx, odd_vals, 'mo-', linewidth=2, markersize=6, label='Нечётные {x_{2n+1}}')
ax3.axhline(y=x_star, color='r', linestyle='--', alpha=0.7, linewidth=2, label=f'x*')
ax3.set_xlabel('Индекс n')
ax3.set_ylabel('x_{2n+1}')
ax3.set_title(f'Нечётная подпоследовательность (монотонно возрастает)')
ax3.grid(True, alpha=0.3)
ax3.legend()

is_increasing = all(odd_vals[i] < odd_vals[i+1] for i in range(len(odd_vals)-1))
ax3.text(0.02, 0.98, f'Монотонно возрастает: {is_increasing}', 
         transform=ax3.transAxes, va='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen' if is_increasing else 'pink', alpha=0.8))

plt.tight_layout()
plt.show()
```

# Normal: Анализ отображения N=2 при r ∈ (1; 3]

## Отображение N=2
x_{n+1} = r x_n (1 - x_n)^2

## 4.1 Неподвижные точки

### Уравнение для неподвижных точек:
x* = r x* (1 - x*)^2

### Решения:
1. x* = 0 (тривиальная точка)

2. Из уравнения:
   1 = r (1 - x*)^2
   (1 - x*)^2 = 1/r
   1 - x* = ±1/√r

   Так как x* ∈ [0, 1], то:
   1 - x* = 1/√r
   x* = 1 - 1/√r

### Итог:
- x*₁ = 0 (всегда существует)
- x*₂ = 1 - 1/√r (существует при r ≥ 1)

## 4.2 Диапазон r для монотонной сходимости к нулю

### Условие монотонной сходимости к нулю:
1. Последовательность {x_n} монотонно убывает
2. lim_{n→∞} x_n = 0

### Анализ монотонности:
Рассмотрим отношение x_{n+1}/x_n:
x_{n+1}/x_n = r (1 - x_n)^2

Для монотонного убывания к нулю нужно:
1. x_{n+1} < x_n
2. x_{n+1}/x_n < 1
3. r (1 - x_n)^2 < 1

### Наихудший случай (максимальное (1 - x_n)^2):
(1 - x_n)^2 достигает максимума 1 при x_n = 0

Таким образом, для гарантированной монотонной сходимости к нулю:
r × 1 < 1
r < 1

### Проверка при x_n → 0:
Если x_n близко к 0, то (1 - x_n)^2 ≈ 1
Тогда x_{n+1}/x_n ≈ r

Для монотонного убывания нужно r < 1

### Итог для 4.2:
Последовательность монотонно сходится к нулю при r < 1
Диапазон: r ∈ [0, 1)

## 4.3 Поведение при r ∈ (1; 3]

При r ∈ (1; 3]:
1. Существуют две неподвижные точки: 0 и 1 - 1/√r
2. Точка 0 неустойчива
3. Точка 1 - 1/√r устойчива при r ∈ (1, 2)
4. При r > 2 могут возникать осцилляции

### Предельное поведение:
lim_{n→∞} x_n = 1 - 1/√r (при подходящих начальных условиях)

```python
import numpy as np
import matplotlib.pyplot as plt

def variant_map(x, r):
    """Отображение N=2: x_{n+1} = r x_n (1 - x_n)^2"""
    return r * x * (1 - x)**2

def plot_trajectories():
    """Построение графиков траекторий для разных r"""
    
    r_values = [0.5, 0.8, 1.2, 1.5, 2.0, 2.5]
    x0 = 0.3
    n_steps = 30
    colors = ['blue', 'green', 'orange', 'red', 'purple', 'brown']
    
    plt.figure(figsize=(15, 10))
    
    for idx, r in enumerate(r_values):
        plt.subplot(2, 3, idx+1)
        
        x_star1 = 0
        x_star2 = 1 - 1/np.sqrt(r) if r >= 1 else None
        
        trajectory = [x0]
        x = x0
        for _ in range(n_steps):
            x = variant_map(x, r)
            trajectory.append(x)
        
        plt.plot(range(n_steps+1), trajectory, 'o-', 
                color=colors[idx], linewidth=2, markersize=4,
                label=f'r = {r}')
        
        plt.axhline(y=x_star1, color='black', linestyle='--', 
                   alpha=0.5, label=f'x*₁ = 0')
        
        if x_star2 is not None:
            plt.axhline(y=x_star2, color='red', linestyle='--',
                       alpha=0.7, label=f'x*₂ = {x_star2:.3f}')
        
        plt.title(f'r = {r}, x₀ = {x0}')
        plt.xlabel('n (шаг)')
        plt.ylabel('x_n')
        plt.grid(True, alpha=0.3)
        plt.legend(fontsize=8)
        plt.ylim(-0.05, 0.8)
    
    plt.suptitle('Отображение N=2: $x_{n+1} = r x_n (1 - x_n)^2$', fontsize=14)
    plt.tight_layout()
    plt.show()

plot_trajectories()
```

## Задание 1. Исследование циклов при r ∈ (3; r∞) в логистическом отображении

### 1. Как изменяется длина цикла при r ∈ (3; r∞)?

При увеличении r от 3 до r∞ ≈ 3.5699456 происходит **каскад бифуркаций удвоения периода**:

- **r = 3.0** → рождается цикл периода 2
- **r = 3.44949** → цикл периода 2 теряет устойчивость, рождается цикл периода 4
- **r = 3.54409** → цикл периода 4 → периода 8
- **r = 3.56441** → цикл периода 8 → периода 16
- **r = 3.56876** → цикл периода 16 → периода 32
- **r → r∞** → бесконечный каскад, длина цикла стремится к бесконечности

**Последовательность длин циклов:** 1 → 2 → 4 → 8 → 16 → 32 → 64 → ...

**Математически:** m(r) = 2^k, где k увеличивается при переходе через точки бифуркации.

### 2. Какие ограничения действуют на m при r ∈ (3; r∞)?

1. **Основные значения m:** m = 2^k (степени двойки)
   - Это основной каскад бифуркаций
   - Наиболее вероятные значения при случайном выборе r

2. **Окна периодичности:** между каскадами существуют интервалы r с другими периодами
   - Период 3: r ≈ 3.8284
   - Период 5: r ≈ 3.7382
   - Период 6: r ≈ 3.6266
   - Период 7: r ≈ 3.7016

3. **Теорема Шарковского:** определяет порядок появления циклов
   - Если есть цикл периода m, то есть циклы всех периодов, больших m в порядке Шарковского

4. **Универсальность:**
   - Постоянная Фейгенбаума δ ≈ 4.669
   - Самоподобие бифуркационной диаграммы
   - Масштабная инвариантность

### Выводы

1. Длина цикла m в основном принимает значения **2^k**
2. Существуют **окна периодичности** с другими m
3. Бифуркации образуют **геометрическую прогрессию**, сходящуюся к r∞
4. Поведение **универсально** для класса одномерных отображений

```python
import numpy as np

def logistic_map(x, r):
    """Логистическое отображение"""
    return r * x * (1 - x)

def find_cycle_period(r, x0=0.3, tol=1e-12):
    """Находит период цикла для заданного r"""
    x = x0
    # Пропускаем переходный процесс
    for _ in range(2000):
        x = logistic_map(x, r)
    
    orbit = []
    for _ in range(1000):
        orbit.append(x)
        x = logistic_map(x, r)
    
    for period in range(1, 100):
        if period_test(orbit, period, tol):
            return period
    return 0

def period_test(orbit, period, tol):
    """Проверяет, является ли период минимальным"""
    if len(orbit) < 2 * period:
        return False
    
    for i in range(period):
        if abs(orbit[-i-1] - orbit[-i-1-period]) > tol:
            return False
    
    for divisor in range(1, period):
        if period % divisor == 0 and period_test(orbit, divisor, tol):
            return False
    
    return True

r_inf = 3.5699456
r_values = np.linspace(3.001, r_inf - 0.001, 1000)
periods_found = set()

print("Анализ периодов для r ∈ (3; r∞)...\n")

for r in r_values:
    period = find_cycle_period(r)
    if period > 0:
        periods_found.add(period)

periods_list = sorted(list(periods_found))

print(f"Найдено различных периодов: {len(periods_list)}")
print("Периоды m:", periods_list[:15], "..." if len(periods_list) > 15 else "")

powers_of_two = []
other_periods = []

for p in periods_list:
    if p > 0 and (p & (p - 1)) == 0:
        powers_of_two.append(p)
    else:
        other_periods.append(p)

print(f"\n1. Степени двойки: {powers_of_two}")
print(f"   Количество: {len(powers_of_two)}")

print(f"\n2. Другие периоды: {other_periods[:10]}", 
      "..." if len(other_periods) > 10 else "")
print(f"   Количество: {len(other_periods)}")

print("\nПримеры для конкретных r:")
test_points = [
    (3.1, "Немного больше 3"),
    (3.45, "Около первой бифуркации"),
    (3.56, "Близко к r∞"),
    (3.83, "Окно периода 3")
]

for r_val, desc in test_points:
    period = find_cycle_period(r_val)
    print(f"r = {r_val:.3f} ({desc}): m = {period}")
```

# Лестница Ламерея для логистического отображения

Рассматриваем логистическую функцию:

\[
f_r(x) = r\,x(1-x)
\]

Лестница Ламерея (cobweb diagram) — графический метод исследования сходимости итерационного процесса.  
Алгоритм:

1. Из точки \((x_n, x_n)\) идём **вертикально** к точке \((x_n, f(x_n))\).
2. Затем из неё — **горизонтально** в точку \((f(x_n), f(x_n))\).
3. Повторяем шаги, используя \(x_{n+1} = f(x_n)\).

Ниже дана функция, которая строит лестницу Ламерея для выбранного параметра \(r\), начального значения \(x_0\) и числа шагов.


```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(x, r):
    return r * x * (1 - x)

def lamerey_staircase(r, x0, steps=20):
    """
    Строит лестницу Ламерея для функции f_r(x) = r x (1 - x).
    r     — параметр отображения
    x0    — начальное значение
    steps — количество итераций
    """

    xs = np.linspace(0, 1, 400)
    ys = logistic(xs, r)

    plt.figure(figsize=(6,5))
    plt.plot(xs, ys)
    plt.plot(xs, xs)

    x = x0
    for _ in range(steps):
        y = logistic(x, r)

        plt.plot([x, x], [x, y])

        plt.plot([x, y], [y, y])

        x = y

    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(f"Лестница Ламерея: r = {r}, x0 = {x0}")
    plt.grid(True)
    plt.show()
```

```python
lamerey_staircase(r=2.5, x0=0.2, steps=25)
lamerey_staircase(r=3.3, x0=0.2, steps=25)
```

Каких-то стабильных уровней больше не наблюдается — порядок полностью теряется.

- **Период-4 и выше:** множество устойчивых уровней; фигура становится сложнее.  
- **Хаос:** отсутствует повторяющаяся структура, траектория заполняет область графика.

g(x_{n+1}) = r\,x_n(1 - x_n)^2

в зависимости от параметра r, построить орбитную диаграмму и график периода.

## 2. Орбитная диаграмма и график периодов
Для каждого значения параметра выполняются итерации, затем определяется период.  
Ниже приведён код расчёта и построений.

---

## 3. Выводы

### • Устойчивая неподвижная точка (период 1)
При малых значениях параметра траектории сходятся к одной точке.

### • Период-2
При увеличении параметра возникает колебание между двумя значениями.

### • Каскад удвоений периода (4, 8, 16, …)
Увеличение параметра приводит к бифуркациям и росту периода.

### • Хаос
Для больших значений параметра период не определяется, траектория нерегулярная.

### • Сравнение с логистическим отображением
Качественно поведение совпадает: стабильная точка → период-2 → удвоение периодов → хаос.  
Отличие — другие критические значения параметра r.


```python
import numpy as np
import matplotlib.pyplot as plt

# Функция отображения
def g(x, r):
    return r * x * (1 - x)**2

# Параметры
r_vals = np.linspace(0, 27/4, 600)
n_transient = 800
n_collect = 800

orbits_r = []
orbits_x = []
periods = []

for r in r_vals:
    x = 0.5

    # Транзиентные итерации
    for _ in range(n_transient):
        x = g(x, r)

    # Сбор данных для анализа цикла
    data = []
    for _ in range(n_collect):
        x = g(x, r)
        data.append(x)

    orbits_r.extend([r]*len(data))
    orbits_x.extend(data)

    # Определение периода
    data_arr = np.array(data[-200:])
    found = None
    for p in range(1, 60):
        if np.allclose(data_arr[:-p], data_arr[p:], atol=1e-5):
            found = p
            break
    periods.append(found)

# График периода
plt.figure(figsize=(12,5))
plt.plot(r_vals, periods, '.', markersize=3)
plt.xlabel("r")
plt.ylabel("Обнаруженный период")
plt.title("Длина устойчивого цикла в зависимости от параметра r")
plt.grid(True, alpha=0.3)
plt.show()

# Орбитная диаграмма
plt.figure(figsize=(10,6))
plt.scatter(orbits_r, orbits_x, s=0.5)
plt.xlabel("r")
plt.ylabel("x")
plt.title("Орбитная диаграмма отображения g(x)=r·x·(1−x)^2")
plt.grid(True, alpha=0.3)
plt.show()
```
